```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

def preprocess_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (256, 256))
    image = image / 255.0
    return image

model_path = r"C:\Users\PC\Documents\optimized_unet_model_checkpoint1.h5"
model = tf.keras.models.load_model(model_path)

input_image_path = r"C:\Users\PC\Desktop\Screenshot 2023-12-13 175444.png"

input_image = preprocess_image(input_image_path)
input_image = np.expand_dims(input_image, axis=0)

predictions = model.predict(input_image)
predictions = np.squeeze(predictions, axis=0)

# Display the images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(input_image[0])
plt.title("Input Image")
plt.axis("off")
plt.savefig(r"C:\Users\PC\Downloads\input_image.png")

plt.subplot(1, 3, 2)
plt.imshow(predictions, cmap='gray')
plt.title("Model Predictions")
plt.axis("off")
plt.savefig(r"C:\Users\PC\Downloads\predictions.png")

plt.subplot(1, 3, 3)
plt.title("No Label Image Required")
plt.axis("off")
plt.savefig(r"C:\Users\PC\Downloads\no_label_image_required.png")

plt.show()
```
